### Задание 1

В магазин привезли устройства с 3-х разных предприятий. 

Соотношение устройств следующее: 20% - продукция первого предприятия, 30% - продукция второго предприятия, 50% - продукция третьего предприятия; далее, 10% продукции первого предприятия высшего сорта, на втором предприятии - 5% и на третьем - 20% продукции высшего сорта. 

Найти вероятность того, что случайно купленная новая продукция окажется высшего сорта.

```python
A - событие, соответствующее покупки продукции первого предпиятия
В - событие, соответствующее покупке продукции второго предприятия
С - событие, соответствующее покупке продукции третьего предприятия
D - событие, соответсвующее покупке продукции высшего сорта
AD - событие, соответствующее получения продукции высшего сорта при условии покупки от первого предприятия
ВD - событие, соответствующее получения продукции высшего сорта при условии покупки от второго предприятия
СD - событие, соответствующее получения продукции высшего сорта при условии покупки от третьего предприятия
E - событие, соответствующее покупке продукции высшего сорта

A,В,С - полная группа событий

p(E) = p(AD) + p(BD) + p(CD)  

p(E) = 0.2*0.1 + 0.3*0.5 + 0.5*0.2 = 0.02 + 0.15 + 0.1 = 0.27
```
