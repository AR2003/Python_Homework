## Задание 1

```python
# Задание 1

import pandas as pd
pd.set_option('display.max_rows', None)

colnames = ['surgery?', 'Age', 'nn1','rectal temperature','pulse','respiratory rate','temperature of extremities','nn2','nn3','nn4','pain','nn5','nn6','nn7','nn8','nn9','nn10','nn11','nn12','nn13','nn14','nn15','outcome','nn16','nn17','nn18','nn19','nn20']
log = pd.read_csv("https://raw.githubusercontent.com/obulygin/pyda_homeworks/master/statistics_basics/horse_data.csv",na_values='?',sep = ',',names=colnames, header=None) 

log.drop(['nn1', 'nn2','nn3','nn4','nn5','nn6','nn7','nn8','nn9','nn10','nn11','nn11','nn12','nn13','nn14','nn15','nn16','nn17','nn18','nn19','nn20'], axis=1, inplace=True)
log.head(10)
```

## Задание 2

```python
print("Максимальные значения")
print(log.max())
print("")
print("Минимальные значения")
print(log.min())
print("")
print("Размах")
print(log.max() - log.min())
print("")
print("Среднее значение")
print(log.mean())
print("")
print("Мода")
print(log['pulse'].mode()[0])
print("")
print("Медиана")
print(log.median())
print("")
print("Среднеквадратичное отклонение")
print("")
print(log.std())
print("")
print("Дисперсия")
print(log.var())
print("")
print("Квантиль 25")
print(log.quantile(0.25))
print("")
print("Квантиль 75")
print(log.quantile(0.75))
print("")
print("Межквартильные размах")
print(log.quantile(0.75) - log.quantile(0.25))
print("")

# одним методом выводим все основные метрики
print(log.describe())
print("")

# рассчитываем выбросы для показателя пульса на базе формулы с межквартильным размахом
q1 = log["pulse"].quantile(0.25)
q3 = log["pulse"].quantile(0.75)
iqr = q3-q1
lower_bound = q1 - 1.5*iqr
upper_bound = q3 + 1.5*iqr
remove_outliers = log[log["pulse"].between(lower_bound,upper_bound,inclusive = True)]

print(lower_bound,upper_bound)
print("")
print("Выбросы")
print(log[-log["pulse"].between(lower_bound,upper_bound,inclusive = True)])

# для примера оценка разницы расчета среднего арифметического для показателя пульс с выбросами и без выбросов
log["pulse"].mean() - remove_outliers["pulse"].mean() 
```

```python
log.head(300)
# Расчет количества пропусков по столбцам
print("Количество пропусков по колонке")
print(log.isna().sum(axis = 0))
print("")
print("Процент пропусков по колонке")
((len(log)-log.count())/len(log))*100
```

```python
# Удаляем строки с количеством пропусков больше двух. Сравниваем статистические показатели, медиана и дисперсия различаются минимально, мешньше 1%
log_6=log.copy()
log_6.dropna(thresh = 6,inplace=True) 
print(log.info())
print(log_6.info())
print(log.describe())
print(log_6.describe())

#print(log.isna().sum(axis = 0))
# удаляем одну строку с незаполненными "surgery?""  и "outcome". Таким образом исключаем по единственному пропуску обоих данных столбцов  
log_6.drop(log_6[log_6["surgery?"].isna()].index,inplace=True)
print(log_6.info())

# заполняем пропуски "rectal temperature" и "pulse" 
log_6_rtp = log_6.copy()
log_6_rtp["rectal temperature"] = log_6_rtp["rectal temperature"].fillna(log_6_rtp["rectal temperature"].median())
log_6_rtp["pulse"] = log_6_rtp["pulse"].fillna(log_6_rtp["pulse"].median())
print(log_6['rectal temperature'].describe())
print(log_6_rtp['rectal temperature'].describe())
```

```python
# примеры поисков закономерностей по связям между колонками для заполнения неизвестных значений с помощью известных - не выявлено,
# варианты - удаление или заполнение медианой,модой или средним
log_6_rtp.groupby(["rectal temperature"])["pain"].median()
log_6_rtp.groupby(["rectal temperature","pulse"])["temperature of extremities"].median()
log_6_rtp["respiratory rate"] = log_6_rtp["respiratory rate"].fillna(log_6["respiratory rate"].median())
log_6_rtp["temperature of extremities"] = round(log_6_rtp["temperature of extremities"].fillna(log_6_rtp["temperature of extremities"].median()))
log_6_rtp["pain"] = round(log_6_rtp["pain"].fillna(log_6_rtp["pain"].median()))
print(log.describe())
print(log.describe())

```
