```python
documents = [
{'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
{'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
{'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
'1': ['2207 876234', '11-2'],
'2': ['10006'],
'3': []
}

def owner_by_number(doc_number):   
    for element in documents:        
        if element['number'] == doc_number:
            print("Владелец документа: ",element['name'])
            return element['name']
    print("Документ не найден в базе")    
    return "Документ не найден в базе"    


def shelf_by_number(doc_number):   
    for key,value in directories.items():        
        if doc_number in value:           
            print("Документ хранится на полке: ",key)
            return key
    print("Документ не найден в базе")    
    return "Документ не найден в базе" 


def document_list():
    all_documents = []
    print("Текущий список документов:") 
    for element in documents:
        shelf = []        
        for key,value in directories.items():        
             if element['number'] in value:                    
                    shelf = key                     
        print(f"№ {element['number']}, тип: {element['type']}, владелец: {element['name']}, полка хранения: {shelf} ")         
    return 


def add_shelf(number):  
    if number in directories:      
        print(f" Такая полка уже существует. Текущий перечень полок: {','.join(list(directories.keys()))}")
    else:
        directories.setdefault(number,[])
        print(f" Полка добавлена. Текущий перечень полок: {','.join(list(directories.keys()))}")    
    return       
 
    

def del_shelf(number):
    deleting_number = directories.get(number,"Not found")      
    if deleting_number == "Not found":
        print(f" Такой полки не существует. Текущий перечень полок: {','.join(list(directories.keys()))}")
    elif len(directories[number]) != 0:
        print(f" На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: {','.join(list(directories.keys()))}")
    else:    
        del(directories[number]) 
        print(f" Полка удалена. Текущий перечень полок: {','.join(list(directories.keys()))}")
    return


def add_document(*params):
    if params[3] not in directories:
        print("Такой полки не существует. Добавьте полку командой as.")
        return
    transaction = {
        'type': params[0],
        'number': params[1],
        'name': params[2]
    }    
    documents.append(transaction)
    directories[params[3]].append(params[1])
    document_list()
    return

def del_document(doc_number):
    for element in documents:        
        if element['number'] == doc_number:
            documents.remove(element)
            key = shelf_by_number(doc_number)
            directories[key].remove(doc_number)
            print("Документ удален")
            document_list()
            return
    print("Документ не найден в базе")
    return


def move_document(doc_number,shelf_number):
    for element in documents:        
        if element['number'] == doc_number:
            if shelf_number in directories: 
                key = shelf_by_number(doc_number)
                directories[key].remove(doc_number)
                directories[shelf_number].append(doc_number)                
                print("Документ перемещен")
                document_list()
                return
            else:
                print(f" Такой полки не существует. Текущий перечень полок: {','.join(list(directories.keys()))}")
                return
    print("Документ не найден в базе")
    return


command = " "
while command != "q":
    command = input("Введите комманду:")
    if command == "p":
        document_number = input("Введите номер документа:")
        owner_by_number(document_number)
    elif command == "s":    
        document_number = input("Введите номер документа:")
        shelf_by_number(document_number)
    elif command == "l":            
        document_list()
    elif command == "as":
        shelf_number = input("Введите номер полки:")
        add_shelf(shelf_number)   
    elif command == "ds":
        shelf_number = input("Введите номер полки:")
        del_shelf(shelf_number)        
    elif command == "ad":
        document_number = input("Введите номер документа:")
        document_type = input("Введите тип документа:")
        document_owner = input("Введите владельца документа:")
        document_shelf = input("Введите полку для хранения:")
        add_document(document_type,document_number,document_owner,document_shelf)
    elif command == "d":
        document_number = input("Введите номер документа:")
        del_document(document_number)
    elif command == "m":
        document_number = input("Введите номер документа:")
        document_shelf = input("Введите номер полки:")
        move_document(document_number,document_shelf)  
            
        
print("Коммандная сессия завершена")
    
```
