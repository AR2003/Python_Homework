```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
df = pd.read_csv("C://water.csv",error_bad_lines=False)
df.drop(["Unnamed: 0"], axis=1,inplace = True)
# строим визуализации mortality-hardness и заодно визуализации зависимости между всеми элементами датасета
df.plot(kind="scatter",x="mortality",y="hardness",title="Точечная диаграмма mortality-hardness")
sns.pairplot(df)
```

```python
# расчет коэффицентов корелляции
print(df[["hardness","mortality"]].corr())
print(df[["hardness","mortality"]].corr(method="spearman"))
print(df[["hardness","mortality"]].corr(method="kendall"))
# расчет показывает среднюю линейную обратную корреляцию между смертностью и жесткостью воды
sns.heatmap(df.corr(),annot=True)
```

```python
X = df[["hardness"]]
y = df["mortality"]
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state = 42)
model=LinearRegression()
model.fit(X_train,y_train)
print(model.coef_)
print(model.intercept_)
y_pred=model.predict(X_test)
```

```python
# модель линейно регрессии
# y = -3.13*X + 1678 
```

```python
# рассчитываем коэффициент детерминации
model.score(X_test,y_test)
```

```python
plt.scatter(X_test,y_test)
plt.plot(X_test,y_pred,c="r")
```

```python
# график остатков
plt.scatter(X_test,y_test-y_pred)
```
